## Стохастический метод

```python
'''
+------------->x
|(0,0)(1,0)...
|(0,1)(1,1)...
|  :    :
|
v
y

'''
#return new_coord, reward
def enviroment(current_coord, move_name):
    move = {'l':np.array([-1, 0]),
            'r':np.array([1, 0]),
            'u':np.array([0, -1]),
            'd':np.array([0, 1]),
           }
    new_coord = current_coord + move[move_name]
    if all(current_coord == (1, 0)):
        return np.array([1, 4]), 10
    elif all(current_coord == (3, 0)):
        return np.array([3, 2]), 5
    elif any(np.concatenate([new_coord < 0, new_coord>4])):
        return current_coord, -1
    else:
        return new_coord, 0
```

```python
class random_agent():
    def __init__(self):
        self.coord = np.random.randint(5, size=2)
        self.V = np.zeros((5,5))
    def action(self):
        return np.random.choice(['l', 'r', 'u', 'd'], 1)[0]
    def update(self, new_coord, reward):
        #Надо поправить формулу
        self.V[tuple(self.coord)] = reward + gamma * self.V[tuple(new_coord)]
        self.coord = new_coord
```

```python
from tqdm import tqdm
```

```python
agent = random_agent()
for _ in tqdm(range(100000000)):
    new_coord, reward = enviroment(agent.coord, agent.action())
    agent.update(new_coord, reward)
```
